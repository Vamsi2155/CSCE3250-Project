```python
# Convert the 'date' column to datetime type for proper sorting
df2['date'] = pd.to_datetime(df2['date'])

# Group by 'location' and get the entry with the maximum date for each country
latest_data = df2.loc[df2.groupby('location')['date'].idxmax()]

# Display the first few rows of the updated dataset
latest_data.head()
```

```python
latest_data.shape
```

```python
# Columns to remove by name
columns_to_remove = [
    'date','new_cases', 'new_cases_smoothed', 'new_deaths', 'new_deaths_smoothed',
    'new_cases_per_million', 'new_cases_smoothed_per_million',
    'new_deaths_per_million', 'new_deaths_smoothed_per_million',
    'weekly_icu_admissions', 'weekly_icu_admissions_per_million',
    'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million',
    'new_tests', 'new_tests_per_thousand', 'new_tests_smoothed',
    'new_tests_smoothed_per_thousand', 'new_vaccinations',
    'new_vaccinations_smoothed', 'new_vaccinations_smoothed_per_million',
    'new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred',
    'extreme_poverty',
     'reproduction_rate', 'icu_patients', 'icu_patients_per_million',
    'hosp_patients', 'hosp_patients_per_million', 'female_smokers', 'male_smokers',
    'handwashing_facilities', 'hospital_beds_per_thousand', 'human_development_index',
    'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
    'excess_mortality', 'excess_mortality_cumulative_per_million', 'total_tests','total_tests_per_thousand','positive_rate','tests_per_case','test_units','total_vaccinations','tests_units','people_vaccinated','people_fully_vaccinated','stringency_index','total_vaccinations_per_hundred','people_vaccinated_per_hundred','total_boosters_per_hundred','total_boosters','people_fully_vaccinated_per_hundred'
]

# Remove the specified columns
data_cleaned = latest_data.drop(columns=columns_to_remove, errors='ignore')

data_cleaned.head(10)
```

```python
# Merge using 'country' from 'countries_data' and 'location' from 'cleaned_data_filtered'
merged_data = pd.merge(df1, data_cleaned, left_on='country', right_on='location', how='inner')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting by grouping by continent
continent_grouped = merged_data.groupby('continent').agg({
    'total_cases_per_million': 'mean',
    'total_deaths_per_million': 'mean',
    'population': 'sum'
}).reset_index()

# Set up the figure for plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Customizing with a color palette
color_palette = "viridis"

# Plot total cases per million by continent
sns.barplot(x='continent', y='total_cases_per_million', data=continent_grouped, ax=axes[0], palette=color_palette)
axes[0].set_title('Average Total Cases per Million by Continent')
axes[0].set_ylabel('Cases per Million')
axes[0].set_xlabel('Continent')

# Add labels on the bars
for bar in axes[0].patches:
    axes[0].text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.1f}',
                 ha='center', va='bottom', color='black', fontsize=9)

# Plot total deaths per million by continent
sns.barplot(x='continent', y='total_deaths_per_million', data=continent_grouped, ax=axes[1], palette=color_palette)
axes[1].set_title('Average Total Deaths per Million by Continent')
axes[1].set_ylabel('Deaths per Million')
axes[1].set_xlabel('Continent')

# Add labels on the bars
for bar in axes[1].patches:
    axes[1].text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.1f}',
                 ha='center', va='bottom', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the visual style
sns.set(style="whitegrid")

# Data preparation for the grouped bar chart
top_countries = merged_data.nlargest(10, 'total_cases')  # Selecting the top 10 countries based on total cases for clarity

# Creating a grouped bar chart
fig, ax = plt.subplots(figsize=(14, 8))
bar_width = 0.35
index = range(len(top_countries['country']))

bar1 = ax.bar(index, top_countries['total_cases'], bar_width, label='Total Cases', color='b')
bar2 = ax.bar([p + bar_width for p in index], top_countries['total_deaths'], bar_width, label='Total Deaths', color='r')

ax.set_xlabel('Country')
ax.set_ylabel('Counts')
ax.set_title('Total COVID-19 Cases and Deaths per Country')
ax.set_xticks([p + bar_width / 2 for p in index])
ax.set_xticklabels(top_countries['country'], rotation=45)
ax.legend()

# Annotate each bar with its value
for rect in bar1 + bar2:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data preparation for horizontal bar charts
top_countries_per_million = merged_data.nlargest(10, 'total_cases_per_million')  # Selecting top countries based on cases per million

# Creating horizontal bar charts for cases and deaths per million
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

# Cases per million
ax1.barh(top_countries_per_million['country'], top_countries_per_million['total_cases_per_million'], color='skyblue')
ax1.set_title('COVID-19 Total Cases per Million by Country')
ax1.set_xlabel('Total Cases per Million')
ax1.set_ylabel('Country')

# Annotate each bar with its value
for index, value in enumerate(top_countries_per_million['total_cases_per_million']):
    ax1.text(value, index, str(round(value, 2)), ha='left', va='center', color='black')

# Deaths per million
ax2.barh(top_countries_per_million['country'], top_countries_per_million['total_deaths_per_million'], color='salmon')
ax2.set_title('COVID-19 Total Deaths per Million by Country')
ax2.set_xlabel('Total Deaths per Million')
ax2.set_ylabel('Country')

# Annotate each bar with its value
for index, value in enumerate(top_countries_per_million['total_deaths_per_million']):
    ax2.text(value, index, str(round(value, 2)), ha='left', va='center', color='black')

plt.tight_layout()
plt.show()
```
